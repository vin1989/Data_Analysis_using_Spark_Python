```python
from pyspark.sql import SparkSession

import warnings

warnings.filterwarnings('ignore')
import pandas as pd
```

```python
spark = SparkSession.builder.appName('data_cleaning').getOrCreate()
```

```python
#df = spark.read.csv('LoanStats_2018Q4.csv',inferSchema=True,header=True,sep=",")

df = spark.read.format('csv')\
.option('header','true')\
.option('inferSchema','true')\
.option('sep',",")\
.load('LoanStats_2018Q4.csv')
```

```python
df = df.select('term','home_ownership','grade','purpose','int_rate', 'addr_state',
               'loan_status','application_type','loan_amnt','emp_length','annual_inc','dti',
               'delinq_2yrs','revol_util','total_acc','dti_joint')
```

```python
df.select('emp_length').distinct().show(10)
```

```python
from pyspark.sql.functions import regexp_extract,regexp_replace
from pyspark.sql.functions import col
```

```python
regex_string = "years|year|\\+|\\<"

df.select(regexp_replace(col('emp_length'),regex_string,'').alias('emp_new_length'),col('emp_length')).show(10)
```

```python
regex_string = "\d+"

df.select(regexp_extract(col('emp_length'),regex_string,0).alias('emp_new_length'),col('emp_length')).show(10)
```

```python
df = df.withColumn('term_cleaned',regexp_replace(col('term'),'months',''))\
     .withColumn('emplen_cleaned',regexp_extract(col('emp_length'),'\d+',0))
```

```python
df.select(['term','term_cleaned','emp_length','emplen_cleaned']).show(10)
```

```python
df.createOrReplaceTempView('cleaned_data')
```

```python
spark.sql('select term_cleaned,emplen_cleaned from cleaned_data').show(10)
```
